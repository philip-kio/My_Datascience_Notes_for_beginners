# Hands on using heart disease data set
After importing  the data and EDA has been performed on it or is been performed. Visualization provides insights like we have stated.
From the .shape method we can see that there are 303 rows and 14 columns. 
In programming there are various method to do a thing. 
With matplotlib there  are two method for plotting. The quick simple method which is using the word plot. 
And the object oriented  method which we have been doing.


```python
hd = pd.read_csv('data/heart-disease.csv')
```

```python
hd.shape
```

```python
hd.isna().sum()
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=hd['age'],y=hd['chol'],c=hd['target'])
ax.set(title = 'Age and cholesterols level impact on Heart disease',
      xlabel='Age',
      ylabel='Cholesterol level')

ax.legend(*scatter.legend_elements(), title='Target');
```

```python
hd.plot(kind='scatter',x= 'age',y= 'chol',c='target')
```

```python
fig, (ax0,ax1)=plt.subplots(nrows=2,
                    ncols=1,
                    figsize=(10,10))
#plot to each different index
scatter= ax0.scatter(x= hd['age'],
                    y= hd['chol'],
                    c= hd['target'])
#customize ax0
ax0.set(title= 'Heart Diesease and Cholesterol Levels',
       xlabel= 'Age',
        ylabel= 'Cholesterol');
#legend 
ax0,ax1.legend(*scatter.legend_elements(), title= 'Targets');

scatter= ax1.scatter(x=hd['age'],
                     y= hd['thalach'],
                     c= hd['target'])

ax1.set(title= 'Heart Diesease and Max Heart rate',
        xlabel= 'Age',
        ylabel= 'Thalach');

#horizontal line:

ax0.axhline(y=hd['chol'].mean(), linestyle='--');
ax1.axhline(y= hd['thalach'].mean(), linestyle='--');

#Add title the figure
fig.suptitle('Heart Diesease Analysis', fontsize=16, fontweight='bold');
```
